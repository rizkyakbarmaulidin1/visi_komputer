```python
# Install OpenCV
!pip install opencv-python-headless

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload the image
uploaded = files.upload()

# Load the uploaded image
img = cv2.imread(list(uploaded.keys())[0])
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Thresholding to get a binary image
_, thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

# Perform morphological operations
kernel = np.ones((5,5), np.uint8)

# Erosion
erosion = cv2.erode(thresh, kernel, iterations=1)

# Dilation
dilation = cv2.dilate(thresh, kernel, iterations=1)

# Opening (erosion followed by dilation)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

# Closing (dilation followed by erosion)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)

# Plotting the results
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

# Original Image
axes[0].imshow(img_rgb)
axes[0].set_title("Original Image")
axes[0].axis('off')

# Grayscale Image
axes[1].imshow(gray, cmap='gray')
axes[1].set_title("Grayscale Image")
axes[1].axis('off')

# Thresholded Image
axes[2].imshow(thresh, cmap='gray')
axes[2].set_title("Thresholded Image")
axes[2].axis('off')

# Erosion Image
axes[3].imshow(erosion, cmap='gray')
axes[3].set_title("Erosion")
axes[3].axis('off')

# Dilation Image
axes[4].imshow(dilation, cmap='gray')
axes[4].set_title("Dilation")
axes[4].axis('off')

# Closing Image
axes[5].imshow(closing, cmap='gray')
axes[5].set_title("Closing")
axes[5].axis('off')

plt.tight_layout()
plt.show()
```
